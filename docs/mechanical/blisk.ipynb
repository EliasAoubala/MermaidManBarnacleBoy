## 2 - Blisk Turbine Profiles

For the turbine design, we intend to use a CNC milling procedure to form the individual turbine elements as illustrated:

Given this will be CNC machined, we aim to keep the depth of these passages to a maximum L/D of 1 (given the maximum reasonably achievable is 3/4 with low cost) - thus improving manufacturability.

For the tolerances for the turbine profiles themselves, it will be challenging to specify a profile tolerance to suppliers while limiting the cost of this 5 axis CNC component.

In this case, we will default to medium tolerances as defined in ISO 2768-1:



As we know that the blade spacing is currently base lined @ 6.53 mm, we must account for the manufacturing tolerances in the design.

In this case, we will treat this dimension as the minimum blade spacing, and will have to increase the spacing according to the manufacturing tolerance ranges.

With the current spec, we have accounted for a 10% increase in blade spacing due to boundary layer effects. We can evaluate for the raw blade spacing as follows:

$$s = s_{bl}/1.1$$

Based on ISO 2768-1 tolerances, we know that the maximum acceptable toelrance is +/-0.2 mm at our dimensional range. Hence we can say that the new nominal blade spacing must be 6.73 mm.

To thus evaluate for the scaling factor, we can then just re-arrange the relationship previously defined

$$N = \frac{6.73}{s}$$

$$N = \frac{6.73 \cdot 1.1}{s_{bl}}$$

$$N = \frac{6.73 \cdot 1.1}{6.53}$$

$$N = \frac{6.73 \cdot 1.1}{6.53}$$

$$N = 1.133$$

This means we need to increase the blade spacing by 13% relative to nominal to account for the tolerances expected from the machining process. To allow for conservatism, we can increase this to 20%, meaning we use a blade spacing factor of 1.2.

As the blade spacing is fixed (by the number of blades), this will just effectively reduce the chord length of the blades that are used and thus allow for a boundary layer change to occur.

## 3 - Shaft Fits

For the shaft fits, we will want a clearnace fit for ease of assembly.

As the diameter of the bore is 15 mm, and the component is so large, we will use the tigher tolerances here to minimise the unbalance induced by the turbine.

In this case, we will specify a fit such that there is zero tolerance at best case, and worst case 100 um tolerance.

| Dimension      | Nominal | Tolerance   |
| -------------- | ------- | ----------- |
| Shaft Diameter | 14.95   | +0.05/-0.05 |
| Blisk Diameter | 15.05   | +0.05/-0.05 |
