> Date:

## 1 - Background

This document encapsulates the sizing of the O-Rings, taking into account the available manufacturing tolerances of the system.
The Trelleborg O-Ring Calculator has been used for the sizing of these O-Rings. The assumed supplier for O-Rings is polymax or simply bearings.

Given the limited time frame for this project, only stock-sizes of O-Rings will be selected.

Based on manufacturing tolerances, a maximum acceptable tolerance of +/- 0.05 mm will be specified.
