## 3 - Final Key Way Selection 

The Final Keyways selected for both the impeller and turbine are as follows:

### 3.1 - Turbine Keyway dimensions

For the turbine, a 4x4 keyway is intended to be used that has a total height of 18 mm from [Accu](https://www.accu.co.uk/parallel-keys/389889-KP-4-4-18-A4).

Based on tolerancing from [Roymech](https://roymech.org/Useful_Tables/Keyways/keyways.html)

The following dimensions are recommended:

| Dimension          | Value | Tolerance  |
| ------------------ | ----- | ---------- |
| Shaft Groove Depth | 2.5   | + 0.1/-0.0 |
| Hub Groove Depth   | 1.8   | + 0.1/-0.0 |
| Radius             | 0.12  | +/-0.04    |
| Shaft Width        | 4     | +0.03/-0.0 |
| Hub Width          | 4.05     | +0.028/-0.02 |

As we can see, these tolerances are a little tight and likely will need to be relaxed - particularly the shaft width and hub width tolerances as those are beyond current manufacturing capabilities which are +/- 0.05. In addition, we will adjust for a larger fillet radius inline with the bearing fits which is 0.3 max which is inline wtih expections for [corner end mill radius](https://www.hgt-endmills.com/product/83980.html).


We will firstly make all our tolerances symettric:

| Dimension          | Value | Tolerance    |
| ------------------ | ----- | ------------ |
| Shaft Groove Depth | 2.55  | + 0.05/-0.05 |
| Hub Groove Depth   | 1.85  | + 0.05/-0.05 |
| Radius             | 0.12  | +0.04/-0.04          |
| Shaft Width        | 4.015     | +0.015/-0.015   |
| Hub Width          | 4.054  | +0.024/-0.024 |

From here, we can pull out the minimum flat width (discounting the radius):

| Dimensions  | Minimum Width |
| ----------- | ------------- |
| Shaft Width | 3.68 mm       |
| Hub Width   | 3.71 mm       |

We can then augment these numbers by applying a maximum fillet radii we get the following:

| Dimensions  | Minimum Width |
| ----------- | ------------- |
| Shaft Width | 4.28 mm       |
| Hub Width   | 4.31 mm       |

From here, we can apply our standard ISO tolerances to get our dimensions for our keyway

| Dimensions         | Nominal Width | Tolerance    |
| ------------------ | ------------- | ------------ |
| Shaft Width        | 4.4 mm        | +0.1/-0.1    |
| Hub Width          | 4.4 mm        | +0.1/-0.1    |
| Radius             | 0.3           | Max          |
| Shaft Groove Depth | 2.55          | + 0.05/-0.05 |
| Hub Groove Depth   | 1.85          | + 0.05/-0.05 |

For the length, we will allow for 1 mm additional spacing on each side to have a total targetted length of 16 mm.
