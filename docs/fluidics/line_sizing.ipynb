## 1 - Background

This document encapsulates the line sizing conducted for the test stand of the Mermaid Man and Barnacle Boy Turbopump

## 2 - Background Theory

For the modelling of the major and minor losses, the standard Fanning Friction factor relationship has been used:

$$\Delta P = \frac{4 f L}{D} \cdot \frac{\rho v^2}{2}$$

Inorder to solve for the Fanning Friction Factor, we can either use the Moody chart directly, or employ the use of the Swamee equation

$$f = \frac{1}{4}\{ \left( \frac{64}{Re} \right)^8 + 9.5 \left[ \ln \left(\frac{\epsilon}{3.7 D} + \frac{5.74}{Re^{0.9}} \right) - \left( \frac{2500}{Re} \right)^6 \right]^{-16} \}^{\frac{1}{8}}$$

Minor Losses can be estimated based on the table from `Perry's Checmical Engineering Handbook`.
